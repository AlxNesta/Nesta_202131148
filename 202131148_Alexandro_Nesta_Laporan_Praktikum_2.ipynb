```python
# import library
# networkx: Digunakan untuk membuat dan memanipulasi struktur graf serta menjalankan algoritma graf.
# matplotlib.pyplot: Digunakan untuk visualisasi grafik.
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Di sini kita membuat objek graf G dan menambahkan node serta edge.
G = nx.Graph()

# Tambahkan node
# nodes adalah daftar lokasi yang akan dijadikan titik dalam graf.
nodes = ['Kost', 'Lokasi 2', 'Lokasi 3', 'Lokasi 4', 'Lokasi 5', 'Lokasi 6', 'Kampus']
G.add_nodes_from(nodes)

# Tambahkan edge dan bobotnya
# edges adalah daftar hubungan antar lokasi dengan bobot yang merepresentasikan jarak atau waktu tempuh.
edges = [('Kost', 'Lokasi 2', 2), ('Lokasi 2', 'Lokasi 4', 2), ('Lokasi 4', 'Lokasi 5', 1), ('Lokasi 5', 'Kampus', 2),
         ('Lokasi 2', 'Kampus', 3), ('Kost', 'Lokasi 3', 3), ('Lokasi 3', 'Lokasi 4', 1), ('Lokasi 4', 'Kampus', 5),
         ('Lokasi 2', 'Lokasi 6', 2), ('Lokasi 6', 'Kampus', 6), ('Lokasi 3', 'Lokasi 6', 2)]
G.add_weighted_edges_from(edges)
```

```python
# Hitung shortest path
shortest_path = nx.shortest_path(G, source='Kost', target='Kampus', weight='weight')

# Beri warna biru pada shortest path
edge_colors = ['blue' if (u, v) in zip(shortest_path, shortest_path[1:]) or (v, u) in zip(shortest_path, shortest_path[1:]) else 'black' for u, v in G.edges()]

# Visualisasikan graf
pos = nx.spring_layout(G)  # Atur posisi node
nx.draw(G, pos, with_labels=True, edge_color=edge_colors)  # Gambar graf
plt.show()  # Tampilkan grafik
```
